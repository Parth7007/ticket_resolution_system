```python
def clean_dataset(df):
    # Get all tag columns present in the dataset
    tag_cols = [col for col in df.columns if col.startswith("tag_") or col == "tag_9"]
    
    # Merge tags into a list column
    df['tags'] = df[tag_cols].values.tolist()
    
    # Remove NaN from tags
    df['tags'] = df['tags'].apply(lambda tags: [t for t in tags if pd.notna(t)])
    
    # Drop unnecessary columns
    cols_to_drop = [col for col in ['queue', 'language', 'version', 'business_type'] if col in df.columns]
    df.drop(columns=cols_to_drop + tag_cols, inplace=True)
    
    # Keep only the necessary columns
    return df[['subject', 'body', 'answer', 'type', 'priority', 'tags']]
```

```python
df1_clean = clean_dataset(df1)
df2_clean = clean_dataset(df2)
df3_clean = clean_dataset(df3)
```

```python
merged_df = pd.concat([df1_clean, df2_clean, df3_clean], ignore_index=True)
```
